```python
#importando a biblioteca pandas
import pandas as pd
```

```python
#Cria uma variavel com os dados do dataset
#Biblioteca Panda lendo arquivo csv
#Parametros: error_bad_lines - Ignora linha com erros
            #sep - define o separador 
df = pd.read_csv("/content/drive/MyDrive/datasets/Gapminder.csv",error_bad_lines=False,sep =";")
```

```python
#visualizando as cinco primeiras linhas
df.head()
```

```python
#alerar nome de colunas
df = df.rename(columns={"country":"Pais","continent":"Continente","year":"Ano","lifeExp":"Expec. de Vida", "pop":"Pop Total","gdpPercap":"PIB"})
df.head()
```

```python
#Cria uma variável Oceania para exibir apenas os países deste continente
oceania = df.loc[df["Continente"]=="Oceania"]
oceania.head()
```
